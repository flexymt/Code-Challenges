```python
dataset = open('day10.txt').read().split('\n')
```

# Question 1 

```python
cycle_list = list()
cycle_list.append(1)
for cycle in dataset:
    if cycle == 'noop':
        cycle_list.append(0)
    if cycle.split()[0] == 'addx':
        value = cycle.split()[1]
        cycle_list.extend([0,int(value)])
        
answer_sum = 0
quantile = [20,60,100,140,180,220]
for quant in quantile:
    answer = sum(cycle_list[:quant])*quant
    answer_sum = answer_sum + answer
answer_sum    
    
    
```
