```python
def s_position():
    for i,sentence in enumerate(data):
        for index,letter in enumerate(sentence):
            if letter == 'S':
                positioning = [i,index]
    return positioning
```

```python
alphabet_numeric  = list(range(1,27))
alphabet_n_map = dict()
for g in range(0,len(alphabet_numeric)):
    alphabet_n_map[alphabet[g]] = alphabet_numeric[g]
    
alphabet_n_map['S'] = 1
alphabet_n_map['None'] = 'None'
alphabet_n_map['E'] = 26
```

```python
def return_letter(position):
    if position[1] < 0 or position[0] < 0 or position[0] > (len(data) - 1) or position[1] > (len(data[0]) - 1) :
        return 'None'
    return data_copy[position[0]][position[1]]

def return_letter_adjusted_grid(position):
    if position[1] < 0 or position[0] < 0 or position[0] > (len(data) - 1) or position[1] > (len(data[0]) - 1) :
        return 'None'
    return data[position[0]][position[1]]
```

```python
data = open('day12.txt').read().split('\n')
data_copy = data.copy()
visited = [[ False for i in range(95)] for i in range(43)]
distance = [[ 9999 for i in range(95)] for i in range(43)]
distance[20][0] = 0
queue = list()
queue.append(s_position())

pred_list = list()
x = 0
while x < 4000:
    
    x+=1
    
    visited[s_position()[0]][s_position()[1]] = True
    
    removed_pos = queue.pop(0)

    right_pos = [removed_pos[0], removed_pos[1] + 1]
    down_pos =  [removed_pos[0] + 1, removed_pos[1]]
    left_pos = [removed_pos[0], removed_pos[1] -1]
    up_pos =  [removed_pos[0] - 1, removed_pos[1]]

    left_letter_score = alphabet_n_map[return_letter(left_pos)]
    right_letter_score = alphabet_n_map[return_letter(right_pos)]
    down_letter_score = alphabet_n_map[return_letter(down_pos)]
    up_letter_score = alphabet_n_map[return_letter(up_pos)]

    left_letter_adj = return_letter_adjusted_grid(left_pos)
    right_letter_adj = return_letter_adjusted_grid(right_pos)
    down_letter_adj = return_letter_adjusted_grid(down_pos)
    up_letter_adj = return_letter_adjusted_grid(up_pos)
    
    current_score = alphabet_n_map[return_letter(s_position())]

    possible_directions = list()
    score_index = 0
    direction_array = ['left','right','down','up']
    current_grid_letters = [left_letter_adj,right_letter_adj,down_letter_adj,up_letter_adj]
    position_array = [left_pos,right_pos,down_pos,up_pos]
    for future_step_score in [left_letter_score,right_letter_score,down_letter_score,up_letter_score]:
        
        i_1, i_2 = position_array[score_index][0],position_array[score_index][1]
        
        if i_1 < 0 or i_2 < 0 or i_1 > 40 or i_2 > 94:
            score_index+=1
            continue
        elif visited[i_1][i_2] == True:
            score_index+=1
            continue
        elif isinstance(future_step_score, str):
            score_index+=1
            continue
        elif current_score + 1 == future_step_score or future_step_score <= current_score:
            possible_directions.append(direction_array[score_index])
        score_index+=1


    position_dict = dict()
    position_dict['right'] = right_pos
    position_dict['down'] = down_pos
    position_dict['left'] = left_pos
    position_dict['up'] = up_pos

    for pos in possible_directions:
        queue.append(position_dict[pos])
        visited[position_dict[pos][0]][position_dict[pos][1]] = True
        pred_list.append(removed_pos)
        distance[position_dict[pos][0]][position_dict[pos][1]] = distance[removed_pos[0]][removed_pos[1]]  + 1 
        
        
    index_1,index_2 = removed_pos[0],removed_pos[1]
    data[index_1] = data[index_1][:index_2] + '.' + data[index_1][index_2+1:]
   
    index_new_1,index_new_2 = queue[0][0],queue[0][1]
    data[index_new_1] =  data[index_new_1][:index_new_2] + 'S' + data[index_new_1][index_new_2+1:]
    
    if s_position() == e_position():
        break

distance[20][72]
```
